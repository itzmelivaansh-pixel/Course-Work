```python
'''
Problem 1 (Part 1)
'''
def gcd(a, b):
    while b!=0:
        remainder=a % b
        a=b
        b=remainder
    return a
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
answer = gcd(num1, num2)
print("The Greatest common divisor is:", answer)
```

```python
'''
Problem 1 (Part 2)
If one or both arguments are negative, the function converts them to positive numbers using the abs()function before computing the GCD, so the result is always a positive value.
'''
```

```python
'''
Problem 2 (Part 1)
'''
def euler_limit(n):
    # convert to float so the division stays decimal
    step=1/float(n)
    # base of the power expression
    base=1+step
    # repeated multiplication instead of using **
    result=1.0
    for _ in range(n):
        result *= base
    return result
# example usage
print(euler_limit(10))
print(euler_limit(100))
print(euler_limit(1000))
```

```python
'''
Problem 2 (Part 2)
'''
#If you call euler_lim(n) with a huge value like 10^{16}, the function will return 1.0.
#it rounds off "1/n" to zero and the base becomes 1.0 and 1.0 raised to any power is 1.
```

```python
'''
Problem 2 (Part 3)
'''
import math
def factorial(k):
    if k==0:
        return 1
    return k*factorial(k - 1)
def euler_sum(n):
    total=0.0
    for k in range(n):
        total+=1/factorial(k)
    return total
print(euler_sum(0))
print(euler_sum(1))
print(euler_sum(2))
print(euler_sum(5))
print(euler_sum(260))
print("True e=",math.exp(1))
```

```python
'''
Problem 2 (Part 4)
'''
def euler_approx(epsilon):
    true_e=math.exp(1)
    n=1
    while True:
        approx=euler_sum(n)
        if abs(approx-true_e)<epsilon:
            return approx
        n+=1
print(euler_approx(0.1))
print(euler_approx(0.01))
print(euler_approx(0.0001))
print(euler_approx(0.000001))
```

```python
'''
Problem 3 (Part 1)
'''
def is_power_of_2(n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if 2**i==n:
                return True
        return False
is_power_of_2(27)
```

```python
'''
Problem 3 (Part 2)
'''
def is_power(b,n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if b**i==n:
                return True
        return False

is_power(9,27)
```
